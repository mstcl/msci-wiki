```python
#| code-fold: true

model_mini = Model(inputs=model.inputs, outputs=model.layers[0].output)

feature_maps1 = model_mini.predict(x1[0], verbose=0)
feature_maps2 = model_mini.predict(x2[0], verbose=0)
i = 2
j = 3
shift = None
for idx in range(i):
    for jdx in range(j):
        ax = plt.subplot(i, j, idx*j+jdx+1)
        ax.set_xticks([])
        ax.set_yticks([])
        if idx == 0:
            plt.imshow(feature_maps1[:shift, :, 0, jdx], cmap='gray')
        else:
            plt.imshow(feature_maps2[:shift, :, 0, jdx], cmap='gray')
```
